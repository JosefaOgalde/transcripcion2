```python
# Paso 2: Definir ruta del archivo original
input_path = "/content/(Audio) Clase 1 cordero.m4a"
output_path = "/content/(Audio) Clase 1 cordero.wav"
```

```python
# Paso 3: Convertir de .m4a a .wav
import ffmpeg

ffmpeg.input(input_path).output(output_path).run()
print(f"Conversión completa: {output_path}")

```

```python
# Paso 4: Transcribir el archivo .wav usando Whisper
import whisper

# Cargar modelo (puedes cambiar "small" por "base" o "tiny" si tienes poca RAM)
model = whisper.load_model("small")

# Transcribir
result = model.transcribe(output_path, language="es")
print("Transcripción:")
print(result["text"])
```

```python
# Paso 5: Guardar transcripción en archivo .txt
transcription_text = result["text"]
output_txt_path = "/content/transcripcion.txt"

with open(output_txt_path, "w", encoding="utf-8") as f:
    f.write(transcription_text)

print(f"Transcripción guardada en: {output_txt_path}")

```
